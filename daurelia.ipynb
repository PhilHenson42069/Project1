```python
stores_df = df[{"GEOID10","NAME10","Urban","PLNG_REGION","LILATracts_1And10","LILATracts_halfAnd10","LA1and10","LAhalfand10","LA1and20","LATracts_half","LATracts1","LATracts10","LATracts20","LowIncomeTracts","POP2010","lapophalf","lapophalfshare","lalowihalf","lalowihalfshare","lapop1","lapop1share","lalowi1","lalowi1share","OHU2010"}]
stores_df
```

```python
seniors_df.loc[seniors_df['LowIncomeTracts'] == 0, 'LowIncomeTracts'] = 'Not Low Income'
seniors_df.loc[seniors_df['LowIncomeTracts'] == 1, 'LowIncomeTracts'] = 'Low Income'
income_groups = seniors_df.groupby('LowIncomeTracts')

lowincome_sum = income_groups['lapop1'].sum()
seniors_sum = (income_groups['laseniors1'].sum() / lowincome_sum) * 100
seniors_sum.plot(kind= 'bar', ylabel= 'low access population (%)', title= 'seniors with low access vs income tract')
plt.show()
```
